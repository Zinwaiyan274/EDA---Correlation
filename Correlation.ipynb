#In this EDA, I gonna explore the Correlation between temperature and cold, hot drinks sale.


```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/myanmards/resource_files/master/correlation.csv')
data.head()
```

There are few methods to calculate correlation in statistics 

*   Pearson
*   Kendall
*   Spearman 

The result of the calculated values may differ a bit due to different mehods





Let's explore each of them

```python
data.corr(method='pearson', min_periods=1)
```

```python
data.corr(method='kendall', min_periods=1)
```

```python
data.corr(method='spearman', min_periods=1)
```

Intuitively, we all know temperature and code, hot drinks sale is correlate. So let's figure it out by calculated numbers aka how they correlate.

```python
data[['Temp', 'Cold_Drink']].corr()
```

```python
data[['Temp', 'Hot_Drink']].corr()
```

So, How about the temperature and snack? Intuitively, it makes sense that temperature and snack sales aren't correlated.
Le's check the calculated statistics if our intuition is correct.

```python
data[['Temp', 'Snacks']].corr()
```

Numbers are far from one and minus one, we can conclude that these two variables has no correlation. What wonderful our intuition is.

This is the visualization of our coorelation.

```python
data.corr().style.background_gradient(cmap='coolwarm').set_precision(3)
```
